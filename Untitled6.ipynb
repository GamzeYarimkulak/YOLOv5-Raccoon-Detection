```python
import yt_dlp as ydl

video_url = 'https://youtu.be/Tc4AhRK5Z9o'

# Video indirilecek dizinizi belirtin
download_path = '/content/'

# YT-DLP ile video indir
ydl_opts = {
    'format': 'best',  # En iyi √ß√∂z√ºn√ºrl√ºkte video indir
    'outtmpl': download_path + '%(title)s.%(ext)s',  # ƒ∞ndirilen video adƒ±nƒ± ayarlayƒ±n
}

with ydl.YoutubeDL(ydl_opts) as ydl:
    ydl.download([video_url])

# Video adƒ± URL'den alƒ±nan video ID'si ile yapƒ±lacak
video_id = video_url.split('/')[-1]  # URL'den video ID'si alƒ±nƒ±r
print(f"Video indirildi: {download_path}{video_id}.mp4")
```
